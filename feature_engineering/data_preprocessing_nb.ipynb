```python
def normalize_cell(x):
    # ---------- 1. NaN ----------
    if pd.isna(x):
        return []

    value = x

    # ---------- 2. Parse stringified list ----------
    if isinstance(x, str):
        x = x.strip()

        if x.upper() == "N/A":
            return []

        if x.startswith("[") and x.endswith("]"):
            try:
                value = ast.literal_eval(x)
            except Exception:
                value = x
        else:
            value = x

    # ---------- 3. If list / nested list ----------
    if isinstance(value, list):
        flat = []
        for item in value:
            if isinstance(item, list):
                flat.extend(item)
            elif item not in (None, "N/A"):
                flat.append(item)

        # clean each element but KEEP list
        cleaned = []
        for i in flat:
            text = str(i)
            text = re.sub(r"(\\n|\n)", " ", text)
            text = re.sub(r"[^a-zA-Z0-9 ]+", " ", text)
            text = re.sub(r"\s+", " ", text)
            text = text.strip().lower()
            if text:
                cleaned.append(text)

        return cleaned

    # ---------- 4. Single string → return as single-item list ----------
    text = str(value)
    text = re.sub(r"(\\n|\n)", " ", text)
    text = re.sub(r"\bN/A\b", "", text, flags=re.IGNORECASE)
    text = re.sub(r"[^a-zA-Z0-9 ]+", " ", text)
    text = re.sub(r"\s+", " ", text)
    text = text.strip().lower()

    return [text] if text else []
```

```python
X = df_final.drop(["matched_score"], axis=1)
```

```python
df_final = X.apply(lambda col: col.map(normalize_cell))
```
